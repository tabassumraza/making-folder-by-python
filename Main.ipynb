```python
import datetime

date = "2023-05-15"
start_time = "09:00"
end_time = "16:00"

calculate_service_charges(date, start_time, end_time)

def calculate_service_charges(date, start_time, end_time):
    hourly_charge = 4  # Hourly charge for the service
    full_day_charge = 30  # Charge for a full day of service
    tax_percentage = 0.1  # 10% tax on total payment
    
    # Convert start and end times to datetime objects
    start_datetime = datetime.datetime.strptime(start_time, "%H:%M")
    end_datetime = datetime.datetime.strptime(end_time, "%H:%M")
    
    # Calculate the duration in hours
    duration = (end_datetime - start_datetime).total_seconds() / 3600
    
    # If the duration is less than 7 hours, charge hourly rate
    if duration < 7:
        total_charge = duration * hourly_charge
    else:
        total_charge = full_day_charge
    
    # Calculate the tax on total payment
    tax_amount = total_charge * tax_percentage
    total_payment = total_charge + tax_amount
    
    # Print the charges and payment details
    print("Date:", date)
    print("Start Time:", start_time)
    print("End Time:", end_time)
    print("Duration:", duration, "hours")
    print("Total Charge:", total_charge, "dollars")
    print("Tax Amount (10%):", tax_amount, "dollars")
    print("Total Payment:", total_payment, "dollars")
```

```python
import os

main_folder = "OOPs"
feature_names = ["Inheritance", "Polymorphism", "Encapsulation", "Abstraction", "Class", "Object"]

if not os.path.exists(main_folder):
    os.mkdir(main_folder)
    print("Main folder 'OOPs' created.")

for feature in feature_names:
    feature_folder = os.path.join(main_folder, feature)
    if os.path.exists(feature_folder):
        print(f"Folder '{feature}' already exists.")
    else:
        os.mkdir(feature_folder)
        print(f"Folder '{feature}' created.")
```
